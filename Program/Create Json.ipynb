# Save Json

```python
import json
from pprint import pprint

data = [{'a' : 1, 'b' : 2, 'c' : 3, 'd' : 4, 'e' : 5 },
        {'a' : 2, 'b' : 3, 'c' : 4, 'd' : 5, 'e' : 6 }    
       ]


with open('data.json', 'w') as outfile:
    for hostDict in data:
        json.dump(hostDict, outfile)
        outfile.write('\n')
```

# Create Json File untuk Citra Artificial

```python
import glob
import os
import json
import numpy as np
from numpy import genfromtxt
from skimage import io, segmentation, color
import matplotlib.pyplot as plt
from skimage.future import graph
import networkx as nx

path1 = 'img/'
path2 = 'labels'
path3 = 'graph/'
imgFile = []
labelFile = []
graphFile = []
dataKu = []


for filename1 in glob.glob(os.path.join(path1, '*.jpg')):
    imgFile.append(filename1)

for filename2 in glob.glob(os.path.join(path2, '*.csv')):
    labelFile.append(filename2)

for filename3 in glob.glob(os.path.join(path3, '*.gpickle')):
    graphFile.append(filename3)
    
for i in range(len(imgFile)):
    data = {'key': 0, 'img': '0', 'label': '0', 'graph': '0'}
    data['img'] = imgFile[i]
    data['label'] = labelFile[i]
    data['graph'] = graphFile[i]
    
    if (i < 50):
        data['key'] = 1
    elif ( i<100 ):
        data['key'] = 2
    elif ( i<150 ):
        data['key'] = 3
    else:
        data['key'] = 4
        
    dataKu.append(data)

with open('dataKu1.json', 'w') as outfile:
for hostDict in dataKu:
    json.dump(hostDict, outfile)
    outfile.write('\n')
```

```python
aList = []

with open('dataKu1.json') as f:
    for line in f:
        j_content = json.loads(line)
        aList.append(j_content)
        
print '\n'.join(map(str,aList))
```

# Create Json File untuk Citra Motif Batik

```python
import glob
import os
import json
import numpy as np
from numpy import genfromtxt
from skimage import io, segmentation, color
import matplotlib.pyplot as plt
from skimage.future import graph
import networkx as nx

path1 = 'img2/'
path2 = 'labels2'
path3 = 'graph2/'
imgFile = []
labelFile = []
graphFile = []
dataKu2 = []


for filename1 in glob.glob(os.path.join(path1, '*.jpg')):
    imgFile.append(filename1)

for filename2 in glob.glob(os.path.join(path2, '*.csv')):
    labelFile.append(filename2)

for filename3 in glob.glob(os.path.join(path3, '*.gpickle')):
    graphFile.append(filename3)
    
for i in range(len(imgFile)):
    data = {'key': '0', 'img': '0', 'label': '0', 'graph': '0'}
    data['img'] = imgFile[i]
    data['label'] = labelFile[i]
    data['graph'] = graphFile[i]
    
    if (i < 60):
        data['key'] = 'grompol'
    elif ( i<120 ):
        data['key'] = 'kawung'
    else:
        data['key'] = 'parang'
        
    dataKu2.append(data)
    
with open('dataKu2.json', 'w') as outfile:
    for hostDict in dataKu2:
        json.dump(hostDict, outfile)
        outfile.write('\n')
```
