# Graph Matching

## Penginputan Images dan Hasil Graph pada Dataset

```python
import glob
import os
import numpy as np
from numpy import genfromtxt
from skimage import io, segmentation, color
import matplotlib.pyplot as plt
from skimage.future import graph
import networkx as nx

path1 = 'img2/'
path2 = 'graph2/'
imgFile = []
graphFile = []


for filename1 in glob.glob(os.path.join(path1, '*.jpg')):
    imgFile.append(filename1)

for filename2 in glob.glob(os.path.join(path2, '*.gpickle')):
    graphFile.append(filename2)
```

```python
from networkx.algorithms import isomorphism

matchedList = []
img = io.imread('img/img051.jpg')
label = segmentation.felzenszwalb(img, min_size=290 )
g1 = graph.rag_mean_color(img, label, connectivity=2, mode='distance')

for i in range(0,len(graphFile)):
    g2 = nx.read_gpickle(graphFile[i])
    gm = isomorphism.GraphMatcher(g1,g2)
    if(gm.is_isomorphic()):
        matchedList.append(imgFile[i])
        
#     if(i==50):
#         print(gm.mapping)
    
# print(matchedList)
for j in range(0,len(matchedList)):
    imgOut = io.imread(matchedList[j])
    print('Image Source: {}'.format(matchedList[j]))
    plt.imshow(imgOut)
    plt.show()
```
