# Halaman Uji Coba Segmentasi

# Segmentasi kelas Ke-1

```python
%matplotlib inline

import glob
import os
import numpy as np
from skimage import io, segmentation, color

path = 'img/'
aFile = []
count = 0

for filename in glob.glob(os.path.join(path, '*.jpg')):
    aFile.append(filename)
    
for i in range(0,25):
    img = io.imread(aFile[i])
    # minsize 100 = 1, 110=2, 120=1, 130=1,140=1, 150
    labels = segmentation.felzenszwalb(img, min_size=150 )
    labels = labels.astype(int)
    print("Felzenszwalb number of segments: {}".format(len(np.unique(labels))))
    if len(np.unique(labels)) != 2:
        count = count + 1
    if(i<9):
        name = 'img00{}.csv'.format(i+1)
    else:
        name = 'img0{}.csv'.format(i+1)
    # print(name)
    path = 'labels/'+name
    np.savetxt(path, labels, delimiter=",")
    
print "\nsegmen tidak sesuai : {}".format(count)
```

# Segmentasi Kelas ke-2

```python
count = 0

for i in range(25,50):
    img = io.imread(aFile[i])
    # minsize 100 = 3, 110=2, 120=3, 130=3,140=2, 150=2 160=2 170=2 180=2 190=2 ... 270 = 0
    labels = segmentation.felzenszwalb(img, min_size=270 )
    labels.astype(int)
    out1 = color.label2rgb(labels, img, kind='overlay')
    print("Felzenszwalb number of segments: {}".format(len(np.unique(labels))))
    if len(np.unique(labels)) != 2:
        count = count + 1
    if(i<9):
        name = 'img00{}.csv'.format(i+1)
    else:
        name = 'img0{}.csv'.format(i+1)
    # print(name)
    path = 'labels/'+name
    np.savetxt(path, labels, delimiter=",")

print "\nsegmen tidak sesuai : {}".format(count)
```

# Show Segmented Image

```python
import glob
import os
import numpy as np
from numpy import genfromtxt
from skimage import io, segmentation, color
import matplotlib.pyplot as plt

img = io.imread('blank.jpg')
path1 = 'img/'
path2 = 'labels/'
imgFile = []
labelFile = []

for filename1 in glob.glob(os.path.join(path1, '*.jpg')):
    imgFile.append(filename1)

for filename2 in glob.glob(os.path.join(path2, '*.csv')):
    labelFile.append(filename2)

for i in range(0,75):
#     img = io.imread(img)
    label = genfromtxt(labelFile[i], delimiter=',')
    out = color.label2rgb(label, img, kind='overlay')
    print(imgFile[i])
    print(labelFile[i])
    print('jumlah region : {}'.format(len(np.unique(label))))
    plt.imshow(out)
    plt.show()
```
