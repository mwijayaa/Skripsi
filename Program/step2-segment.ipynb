# Halaman Uji Coba Segmentasi

# Segmentasi kelas Ke-1

```python
import glob
import os
import numpy as np
from skimage import io, segmentation, color

path = 'img/'
aFile = []
count = 0

for filename in glob.glob(os.path.join(path, '*.jpg')):
    aFile.append(filename)
    
for i in range(0,50):
    img = io.imread(aFile[i])
    # minsize 100 = 1, 110=2, 120=1, 130=1,140=1, 150
    labels = segmentation.felzenszwalb(img, min_size=290 )
    labels = labels.astype(int)
    print("Felzenszwalb number of segments: {}".format(len(np.unique(labels))))
    if len(np.unique(labels)) != 2:
        count = count + 1
    if(i<9):
        name = 'img00{}.csv'.format(i+1)
    else:
        name = 'img0{}.csv'.format(i+1)
    # print(name)
    path = 'labels/'+name
    np.savetxt(path, labels.astype(int), fmt='%i', delimiter=",")
    
print "\nsegmen tidak sesuai : {}".format(count)
```

# Segmentasi Kelas ke-2

```python
count = 0

for i in range(50,100):
    img = io.imread(aFile[i])
    # minsize 100 = 3, 110=2, 120=3, 130=3,140=2, 150=2 160=2 170=2 180=2 190=2 ... 270 = 0
    labels = segmentation.felzenszwalb(img, min_size=290 )
    labels.astype(int)
    out1 = color.label2rgb(labels, img, kind='overlay')
    print("Felzenszwalb number of segments: {}".format(len(np.unique(labels))))
    if len(np.unique(labels)) != 3:
        count = count + 1
    if(i<9):
        name = 'img00{}.csv'.format(i+1)
    elif(i<99):
        name = 'img0{}.csv'.format(i+1)
    else:
        name = 'img{}.csv'.format(i+1)
    # print(name)
    path = 'labels/'+name
    np.savetxt(path, labels.astype(int), fmt='%i', delimiter=",")

print "\nsegmen tidak sesuai : {}".format(count)
```

# Show Segmented Image

```python
import glob
import os
import numpy as np
from numpy import genfromtxt
from skimage import io, segmentation, color
import matplotlib.pyplot as plt

img = io.imread('blank.jpg')
path1 = 'img/'
path2 = 'labels/'
imgFile = []
labelFile = []

for filename1 in glob.glob(os.path.join(path1, '*.jpg')):
    imgFile.append(filename1)

for filename2 in glob.glob(os.path.join(path2, '*.csv')):
    labelFile.append(filename2)

for i in range(0,200):
#     img = io.imread(img)
    label = genfromtxt(labelFile[i], delimiter=',')
    out = color.label2rgb(label, img, kind='overlay')
    print(imgFile[i])
    print(labelFile[i])
    print('jumlah region : {}'.format(len(np.unique(label))))
    plt.imshow(out)
    plt.show()
```

# Segmentasi pada Motif Batik

```python
import glob
import os
import numpy as np
from skimage import io, segmentation, color

path = 'img2/'
bFile = []

for filename in glob.glob(os.path.join(path, '*.jpg')):
    bFile.append(filename)
    
for i in range(len(bFile)):
    img = io.imread(bFile[i])
    labels = segmentation.felzenszwalb(img, min_size=290 )
    labels = labels.astype(int)
    print("Felzenszwalb number of segments: {}".format(len(np.unique(labels))))
    if(i<9):
        name = 'batik00{}.csv'.format(i+1)
    elif(i<99):
        name = 'batik0{}.csv'.format(i+1)
    else:
        name = 'batik{}.csv'.format(i+1)
    path = 'labels2/'+name
    np.savetxt(path, labels.astype(int), fmt='%i', delimiter=",")
```
