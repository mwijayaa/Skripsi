# Hasil Uji Coba Pembangunan RAG

```python
import glob
import os
import numpy as np
from numpy import genfromtxt
from skimage import io
import matplotlib.pyplot as plt
from skimage.future import graph
import networkx as nx

path1 = 'img/'
path2 = 'labels/'
path3 = 'graph/'
imgFile = []
labelFile = []
graphFile = []

for filename1 in glob.glob(os.path.join(path1, '*.jpg')):
    imgFile.append(filename1)

for filename2 in glob.glob(os.path.join(path2, '*.csv')):
    labelFile.append(filename2)

for i in range(0,75):
    img = io.imread(imgFile[i])
    label = genfromtxt(labelFile[i], delimiter=',')
    label = label.astype(int)
    g =  graph.rag_mean_color(img, label, connectivity=1, mode='distance')
    gshow = graph.show_rag(label,g,img,edge_cmap='viridis',img_cmap='magma',edge_width=3)
    print(imgFile[i])
    print(labelFile[i])
    print('jumlah region : {}'.format(len(np.unique(label))))
    if(i<9):
        name = 'img00{}.gpickle'.format(i+1)
    elif (i<99):
        name = 'img0{}.gpickle'.format(i+1)
    else:
        name = 'img0{}.gpickle'.format(i+1)
    path3 = 'graph/'+name
    nx.write_gpickle(g, path3)
    plt.show()
```
